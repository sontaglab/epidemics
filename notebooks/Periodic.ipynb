# Social distancing with periodic relaxation

## Objective
This notebook is dedicated to understand how social distancing will effect the spread of an epidemic.
## Setup

```julia
using DifferentialEquations, Plots, Plots.PlotMeasures
pyplot()
```

## Utility functions

```julia
# change plot fonts
bplot(p, s) = plot!(p, titlefont=font(s, "Times"), xguidefontsize=s-4
    , ytickfont=font(s-4, "Times"), margin=8mm, yguidefontsize=s-4
    , xtickfont = font(s-4, "Times"), legendfont = font(s-4, "Times"))

# this function will generate a T period of social distancing at time t0
sinput(t, ts, td) = if (t-ts<0 || t-ts-td>0) 0 else 1 end

# this function will generate a periodic input with basic properties
# r the ratio between on and off state should be \in [0,1]
# T period of the input
# t0 start time of the policy
pinput(t, r, T) = if t<0 0 
    elseif r<=0 0 elseif r>=1 1
    elseif T<=0 0
    else 1-sign(floor(rem((t)/T,1)/r)) end

# this function will generate a periodic input with a single pulse
# r the ratio between on and off state should be \in [0,1]
# T period of the input
# ts start time of the strict social distancing policy
# td duration of the social distancing
spinput(t, ts, td, r, T) = if (t-ts>0 && t-ts-td<0) 1 
    elseif r<=0 0 elseif r>=1 1
    elseif T<=0 0
else 1-sign(floor(rem(t/T,1)/r)) end
```

```julia
t = 0:0.01:70
ts, td, T, r = [14, 28, 7, 5/7]

plot(t, pinput.(t, r, T), linewidth=4, label="Periodic")
plot!(t, spinput.(t, ts, td, r, T), linewidth=4
    , label="Periodic and Single Interval", linestyle=:dash)
plot!(t, sinput.(t,ts,td), linewidth=4
    , label="Single Interval", linestyle=:dot)
```

# $SIR$ Model

## Periodic input

```julia
# SIR model description
function SIR1(dx,x,p,t)
    b, db, g, r, T, t0 = p
    u = b - db.*pinput(t, r, T, t0)
    dx[1] = - u*x[1]*x[2]
    dx[2] = u*x[1]*x[2] - g*x[2] 
    dx[3] = g*x[2] 
end

# define key parameters
mainT, main0, mainr = [10.0, 0.0, 0.5]
gamma, beta, betad = [0.05, 0.2, 0.1]
bdiff = beta-betad
tspan = (0.0, 1e3)
u0 = [1-1e-3, 1e-3, 0.]
p  = [beta, bdiff, gamma, 0.0, mainT, main0]

# ode solver
alg = BS3()

Trng = 1:120
rrng = 0:1e-2:1
data = zeros(size(rrng,1), size(Trng,1))

for i = 1:size(rrng,1)
    for j = 1:size(Trng,1)
        # solve the ode problem
        p[4:5] = [rrng[i], Trng[j]]
        prob = ODEProblem(SIR1, u0, tspan, p)
        sol  = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=0.1)
        # save the maximum value of the infected peak
        data[i,j] = maximum(sol[2,:])
    end
end

p1 = heatmap(Trng, 100*rrng, data,c=cgrad([:white, :blue]),
    ylabel="Social Distancing Rate [%]", xlabel="Social Distancing Period [days]",
    title="Max(I)")
p1 = bplot(p1, 16)
savefig(p1, "figures/SIRp_heatmap.png")
savefig(p1, "figures/SIRp_heatmap.svg")

p2 = contour(Trng, 100*rrng, data, ylabel="Social Distancing Rate [%]"
    , title="Max(I)", xlabel="Social Distancing Period [days]"
    , contour_labels = true, linewidth=1, colorbar = false, c=:black)
p2 = bplot(p2, 16)
savefig(p2, "figures/SIRp_contour.png")
savefig(p2, "figures/SIRp_contour.svg")

fig = plot(p1, p2, layout=(1,2), size=(960,470))

savefig(fig, "figures/SIRp.png")
savefig(fig, "figures/SIRp.svg")
```

# Periodic Input with Single Pulse

```julia
# SIR model description
function SIR2(dx,x,p,t)
    b, db, g, ts, td, r, T = p
    u = b - db.*spinput(t, ts, td, r, T)
    dx[1] = - u*x[1]*x[2]
    dx[2] = u*x[1]*x[2] - g*x[2] 
    dx[3] = g*x[2] 
end

# define key parameters
mainT, mainr, ts, td = [7., 5/7, 0, 5]
gamma, beta, betad = [0.05, 0.2, 0.1]
bdiff = beta-betad
tspan = (0.0, 1e3)
u0 = [1-1e-3, 1e-3, 0.]
p  = [beta, bdiff, gamma, ts, td, mainr, mainT]

p[4:5] = [0, 0]
prob  = ODEProblem(SIR2, u0, tspan, p)
sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=1e-2)
timen = sol.t[argmax(sol[2,:])] # time of the peak
p[4:5] = [0, 1e3]
prob  = ODEProblem(SIR2, u0, tspan, p)
sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=1e-2)
timed = sol.t[argmax(sol[2,:])] # time of the peak

display([timen, timed])

# range of the quarantine starting time and duration
Trng = 0:160
trng = 0:110
data = zeros(size(Trng,1), size(trng,1))
time = data



for i = 1:size(Trng,1)
    for j = 1:size(trng,1)
        # solve the ode problem
        p[4:5] = [trng[j], Trng[i]]
        prob = ODEProblem(SIR2, u0, tspan, p)
        sol  = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=0.1)
        data[i,j] = maximum(sol[2,:])
    end
end

p1 = heatmap(trng, Trng, data,c=cgrad([:white, :blue]),
    xlabel="SD Start Time ts [days]", ylabel="SD Duration td [days]",
    title="Max(I)")
p1 = plot!(p1, [Int(round(timen))], seriestype="vline", line=:dash, linewidth=6, legend=:false)
p1 = bplot(p1, 16)

savefig(p1, "figures/SIRsp_heatmap.png")
savefig(p1, "figures/SIRsp_heatmap.svg")
```

# Weighted Average $\bar \beta$ with Single Pulse

```julia
# model description
function SIR(dx,x,p,t)
    b, db, g, ts, td = p
    u = b - db.*sinput(t, ts, td)
    dx[1] = -u*x[1]*x[2]
    dx[2] = u*x[1]*x[2] - g*x[2]
    dx[3] = g*x[2]
end

beta  = 0.2*2/7+0.1*5/7
bdiff = beta-0.1
gamma = 0.05

# define initial values
u0 = [1-1e-3, 1e-3, 0.]
p  = [beta, bdiff, gamma, 0, 0]
tspan = (0.0, 1e3)

prob  = ODEProblem(SIR, u0, tspan, p)
sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=1e-2)
timen = sol.t[argmax(sol[2,:])]
p[4:5] = [0, 1e3]
prob  = ODEProblem(SIR, u0, tspan, p)
sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=1e-2)
timed = sol.t[argmax(sol[2,:])]

display([timen, timed])

# range of the quarantine starting time and duration
Trng = 0:160
trng = 0:110
data = zeros(size(Trng,1), size(trng,1))
time = data

for i = 1:size(Trng,1)
    for j = 1:size(trng,1)
        # solve the ode problem
        p[4:5] = [trng[j], Trng[i]]
        prob = ODEProblem(SIR, u0, tspan, p)
        sol  = solve(prob, alg, reltol=1e-8, abstol=1e-8, saveat=0.1)
        data[i,j] = maximum(sol[2,:])
    end
end

p = heatmap(trng, Trng, data,c=cgrad([:white, :blue]),
    xlabel="SD Start Time ts [days]", ylabel="SD Duration td [days]",
    title="Max(I)")
p = plot!(p, [Int(round(timen))], seriestype="vline", line=:dash, linewidth=6, legend=:false)
p = bplot(p, 16)
savefig(p, "figures/SIRspaverage_heatmap.png")
savefig(p, "figures/SIRspaverage_heatmap.svg")
```
