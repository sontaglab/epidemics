# Robustness

Robustness analysis for the paper:


Sadeghi, M., Greene, J. and Sontag, E., 2020. Universal features of epidemic models under social distancing guidelines. bioRxiv.


## Setup

```julia
using DifferentialEquations, Plots, Plots.PlotMeasures, LaTeXStrings
pyplot()
```

## Utility functions

```julia
# change plot fonts
bplot(p, s) = plot!(p, titlefont=font(s, "Times"), xguidefontsize=s-4
    , ytickfont=font(s-4, "Times"), margin=8mm, yguidefontsize=s-4
    , xtickfont = font(s-4, "Times"), legendfont = font(s-4, "Times"))

# this function will generate a T period of social distancing at time t0
sinput(t, ts, td) = if (t-ts<0 || t-ts-td>0) 0 else 1 end
```

# Robustness

```julia
# model description
function SIR(dx,x,p,t)
    βn, βd, γ, ts, td = p
    u = βn - (βn-βd).*sinput(t, ts, td)
    dx[1] = -u*x[1]*x[2]
    dx[2] = u*x[1]*x[2] - γ*x[2]
    dx[3] = γ*x[2]
end

function robustness(βn,βd,γ)
    u0 = [1-1e-3, 1e-3, 0.]                               #initial condition
    p  = [βn, βd, γ, 0.0, 0.0]                            #initial prameters
    
    alg = BS3()                                                 # ode solver
    tspan = (0.0, 500.0)                                      # virtual peak 
    prob  = ODEProblem(SIR, u0, tspan, p)
    sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8,saveat=0.1)
    timen = sol.t[argmax(sol[2,:])]    # infected peak for normal population
    p[5]  = 1000
    prob  = ODEProblem(SIR, u0, tspan, p)
    sol   = solve(prob, alg, reltol=1e-8, abstol=1e-8,saveat=0.1)
    timed = sol.t[argmax(sol[2,:])] # infected peak for distanced population
    
    Trng = 0:Int(floor(timed*1.2))               # computational range of td
    trng = 0:Int(floor(timen*1.2))               # computational range of ts
    data = zeros(size(Trng,1), size(trng,1))
    time = data
    
    for i = 1:size(Trng,1)
        for j = 1:size(trng,1)
            # solve the ode problem
            p[4:5] = [trng[j], Trng[i]]
            prob = ODEProblem(SIR, u0, tspan, p)
            sol  = solve(prob, alg, reltol=1e-8, abstol=1e-8,saveat=0.1)
            data[i,j] = maximum(sol[2,:])
        end
    end
    
    p1 = heatmap(trng, Trng, data,c=cgrad([:white, :blue]),
    xlabel="SD Start Time ts [days]", ylabel="SD Duration td [days]",
    title="Max(I)")
    p1 = plot!(p1, [Int(round(timen))], seriestype="vline", line=:dash, 
        linewidth=6, legend=:false)
    p1 = bplot(p1, 16)
    savefig(p1, "figures/robustness_"*string(βn,βd,γ)*"_heatmap.svg")
    savefig(p1, "figures/robustness_"*string(βn,βd,γ)*"_heatmap.png")
    
    p2 = contour(trng, Trng, data,c=cgrad([:white, :blue]), 
        colorbar = false, xlabel="SD Start Time ts [days]", 
        ylabel="SD Duration td [days]", title="Max(I)", 
        contour_labels = true, linewidth=1, 
        levels=range(round(minimum(data),digits=2),
        stop=round(maximum(data),digits=2),length=8))
    s = -(βn-γ)/(βd-γ); ss = Int(floor(timen*0.9));
    x = 2:ss; y = s*x.-s*ss;
    p2 = plot!(p2, x, y, linewidth=6, line=:dash, legend=:false)
    p2 = bplot(p2, 16)
    savefig(p2, "figures/robustness_"*string(βn,βd,γ)*"_contour.svg")
    savefig(p2, "figures/robustness_"*string(βn,βd,γ)*"_contour.png")
    
    fig = plot(p1, p2, layout=(1,2), size=(960,470))
    savefig(fig, "figures/robustness_"*string(βn,βd,γ)*".svg")
    savefig(fig, "figures/robustness_"*string(βn,βd,γ)*".png")
end
```
